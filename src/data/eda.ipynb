```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import warnings
import os
 
warnings.filterwarnings("ignore")
```

```python
# ─── CONFIGURACIÓN ────────────────────────────────────────────────────────────
FILE_PATH = r"..\..\data\interim\Libros_Unificados_Recomendador.xlsx" 
OUTPUT_DIR = "eda_outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)
 
PALETTE = sns.color_palette("Blues_d", 8)
sns.set_theme(style="whitegrid", palette=PALETTE)
plt.rcParams.update({"figure.dpi": 130, "font.family": "DejaVu Sans"})
```

```python
# ─── CARGA ─────────────────────────────────────────────────────────────────────
print("=" * 60)
print("  EDA — BASE DE DATOS LIBROS (post-unificación OpenAlex)")
print("=" * 60)
 
df = pd.read_excel(FILE_PATH)
print(f"\n✔  Dataset cargado: {df.shape[0]:,} filas × {df.shape[1]} columnas\n")
```

```python
df.head()
```

```python
print("━" * 50)
print("1. TIPOS DE DATOS Y VALORES NO NULOS")
print("━" * 50)
print(df.info())
```

```python
print("\n" + "━" * 50)
print("2. COMPLETITUD POR COLUMNA (%)")
print("━" * 50)
completitud = (df.notna().sum() / len(df) * 100).sort_values()
print(completitud.round(1).to_string())
 
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.barh(completitud.index, completitud.values,
               color=[PALETTE[6] if v >= 60 else PALETTE[2] for v in completitud.values])
ax.axvline(60, color="red", linestyle="--", linewidth=1, label="Umbral 60%")
ax.set_xlabel("% de registros con valor")
ax.set_title("Completitud por columna", fontsize=14, fontweight="bold")
ax.legend()
for bar, val in zip(bars, completitud.values):
    ax.text(val + 0.5, bar.get_y() + bar.get_height() / 2,
            f"{val:.1f}%", va="center", fontsize=8)
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/01_completitud_columnas.png")
plt.close()
print(f"\n  → Gráfico guardado: {OUTPUT_DIR}/01_completitud_columnas.png")
```

```python
print("\n" + "━" * 50)
print("3. DISTRIBUCIÓN DE Metodo_Match")
print("━" * 50)
metodo_counts = df["Metodo_Match"].value_counts()
print(metodo_counts.to_string())
 
# ¿Cuántos hicieron match real?
NO_MATCH_LABEL = "no encontrado"  
match_mask = df["Metodo_Match"].str.lower() != NO_MATCH_LABEL
n_match    = match_mask.sum()
n_no_match = (~match_mask).sum()
pct_match  = n_match / len(df) * 100
 
print(f"\n  Con match    : {n_match:,}  ({pct_match:.1f}%)")
print(f"  Sin match    : {n_no_match:,}  ({100 - pct_match:.1f}%)")
print(f"\n  ⚠  Registros disponibles para entrenamiento: {n_match:,}")
 
fig, ax = plt.subplots(figsize=(10, 8))
 
# Barras por método
metodo_counts.plot(kind="bar", ax=ax, color=PALETTE[4::-1], edgecolor="white")
ax.set_title("Frecuencia por Metodo_Match", fontweight="bold")
ax.set_xlabel("")
ax.set_ylabel("Registros")
ax.tick_params(axis="x", rotation=30)
for p in ax.patches:
    ax.annotate(f"{int(p.get_height()):,}",
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha="center", va="bottom", fontsize=9)
 
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/02_metodo_match.png")
plt.close()
print(f"  → Gráfico guardado: {OUTPUT_DIR}/02_metodo_match.png")
```

```python
print("\n" + "━" * 50)
print("4. COMPLETITUD EN REGISTROS CON MATCH")
print("━" * 50)
df_match = df[match_mask].copy()
comp_match = (df_match.notna().sum() / len(df_match) * 100).sort_values()
print(comp_match.round(1).to_string())
 
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(comp_match.index, comp_match.values,
        color=[PALETTE[6] if v >= 60 else PALETTE[2] for v in comp_match.values])
ax.axvline(60, color="red", linestyle="--", linewidth=1, label="Umbral 60%")
ax.set_xlabel("% de registros con valor")
ax.set_title(f"Completitud — solo registros con match (n={n_match:,})",
             fontsize=13, fontweight="bold")
ax.legend()
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/03_completitud_match.png")
plt.close()
print(f"  → Gráfico guardado: {OUTPUT_DIR}/03_completitud_match.png")
```

```python
print("\n" + "━" * 50)
print("5. CAMPOS CLAVE PARA ENTRENAMIENTO DEL MODELO")
print("━" * 50)
campos_clave = ["Abstract", "Keywords", "BISAC Catálogo", "OpenAlex_Concepts",
                "Area_Conocimiento", "W_Editorial", "W_Citas","Idioma","Año de publicación"]
campos_disponibles = [c for c in campos_clave if c in df.columns]
 
resumen = pd.DataFrame({
    "Total (%)":       (df[campos_disponibles].notna().sum() / len(df) * 100).round(1),
    "Con match (%)":   (df_match[campos_disponibles].notna().sum() / len(df_match) * 100).round(1),
})
print(resumen.to_string())
 
fig, ax = plt.subplots(figsize=(10, 6))
resumen.plot(kind="barh", ax=ax, color=[PALETTE[2], PALETTE[6]], edgecolor="white")
ax.set_title("Completitud campos clave — Total vs Con Match", fontweight="bold")
ax.set_xlabel("% registros con valor")
ax.axvline(60, color="red", linestyle="--", linewidth=1)
plt.tight_layout()
plt.savefig(f"{OUTPUT_DIR}/04_campos_clave.png")
plt.close()
print(f"  → Gráfico guardado: {OUTPUT_DIR}/04_campos_clave.png")
```

```python
print("\n" + "━" * 50)
print("6. DISTRIBUCIÓN POR IDIOMA")
print("━" * 50)
if "Idioma" in df.columns:
    idioma_counts = df["Idioma"].value_counts()
    print(idioma_counts.to_string())
 
    fig, ax = plt.subplots(figsize=(8, 5))
    idioma_counts.head(10).plot(kind="bar", ax=ax, color=PALETTE[5::-1], edgecolor="white")
    ax.set_title("Distribución por idioma (top 10)", fontweight="bold")
    ax.set_xlabel("")
    ax.tick_params(axis="x", rotation=30)
    plt.tight_layout()
    plt.savefig(f"{OUTPUT_DIR}/05_idioma.png")
    plt.close()
    print(f"  → Gráfico guardado: {OUTPUT_DIR}/05_idioma.png")
```

```python
print("\n" + "━" * 50)
print("7. DETECCIÓN DE DUPLICADOS")
print("━" * 50)
n_dup_codigo = df.duplicated(subset=["Código del libro"]).sum()
n_dup_titulo = df.duplicated(subset=["Titulo"]).sum()
print(f"  Duplicados por Código del libro : {n_dup_codigo}")
print(f"  Duplicados por Título           : {n_dup_titulo}")
```

```python
print("\n" + "=" * 60)
print("  RESUMEN EJECUTIVO")
print("=" * 60)
print(f"  Total registros            : {len(df):,}")
print(f"  Registros con match        : {n_match:,} ({pct_match:.1f}%)")
print(f"  Registros sin match        : {n_no_match:,} ({100-pct_match:.1f}%)")
print(f"  Métodos de match distintos : {df['Metodo_Match'].nunique()}")
print(f"  Similarity promedio (match): {df_match['OpenAlex_Similarity'].mean():.4f}")
print(f"  Similarity mediana (match) : {df_match['OpenAlex_Similarity'].median():.4f}")
print(f"  Registros con Abstract     : {df['Abstract'].notna().sum():,} "
      f"({df['Abstract'].notna().mean()*100:.1f}%)")
print(f"  Registros con Abstract+match: "
      f"{df_match['Abstract'].notna().sum():,} "
      f"({df_match['Abstract'].notna().mean()*100:.1f}%)")
print(f"\n  Gráficos exportados en   : ./{OUTPUT_DIR}/")
print("=" * 60)
```
